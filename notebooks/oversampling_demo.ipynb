```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
import matplotlib.pyplot as plt
```

```python
num_samples = 100
num_classes = 3
input_shape = (2,)
batch_size = 1

x = np.random.rand(num_samples, *input_shape)
y = np.random.randint(low=0, high=num_classes, size=num_samples)
y = to_categorical(y)
```

```python
# define oversampling strategy
oversampler = RandomOverSampler(sampling_strategy='minority', shrinkage=0.3)
```

```python
# fit and apply the transform
X_over, y_over = oversampler.fit_resample(x, y)
```

```python
y_labels = np.argmax(y, axis=1).tolist()
y_over_labels = np.argmax(y_over, axis=1).tolist()
print("unbalanced: ",sorted(Counter(y_labels).items()))
print("oversampled minority classes: ",sorted(Counter(y_over_labels).items()))
```

```python
color_mappping = {
    0: "green",
    1: "red",
    2: "blue"
}
plt.scatter(x[:,0], x[:,1], color=[color_mappping[label] for label in y_labels])
plt.show()
```

```python
plt.scatter(X_over[:,0], X_over[:,1], color=[color_mappping[label] for label in y_over_labels])
plt.show()
```
