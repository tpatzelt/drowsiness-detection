```python
import pandas as pd
from drowsiness_detection.data import create_eye_closure_karolinksa_dataset
import matplotlib.pyplot as plt
```

```python
data = next(create_eye_closure_karolinksa_dataset())["eye_closure"].to_numpy()
```

```python
methods = ['linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh', 'spline', 'polynomial', 'from_derivatives', 'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']
for method in methods:
    e = pd.DataFrame(data.copy()[900:1000])
    try:
        e_interpolated = e.interpolate(method)
    except Exception as e:
        print(method + " didnt work")
        continue
    e_interpolated.plot()
    plt.title(method)
    plt.axvline(51)
    plt.axvline(61)
    plt.show()
```
