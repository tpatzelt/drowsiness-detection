```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.precision", 2)
from pathlib import Path
from tqdm import tqdm
```

```
feature_df = pd.read_pickle("/data/workspace_files/all_session.pkl")
```

```
# add perclos column
PERCLOS_THRESHOLD = .8
feature_df["perclos_closed_combined"] = feature_df["combined_eye_closure"] >= PERCLOS_THRESHOLD
```

```
def add_perclos(df: pd.DataFrame, interval_in_sec: int = 60, fps: int = 30):
    df["perclos_60s_interval_combined"] = pd.NA
    # df["perclos_60s_interval_combined"] = df["perclos_60s_interval_combined"].astype(pd.Int64Dtype())
    num_interval_frames = interval_in_sec * fps
    for top_index in feature_df.index.get_level_values('filename'):
        print("processing top-index: ", top_index)
        sub_df = df.loc[top_index]
        for index in tqdm(sub_df.index[num_interval_frames:]):
            s = sub_df.iloc[index-num_interval_frames:index]["perclos_closed_combined"].sum()
            df.loc[(top_index, index),"perclos_60_interval_combined"] = sub_df.iloc[index-num_interval_frames:index]["perclos_closed_combined"].sum()
        break
    
add_perclos(df=feature_df)
feature_df["perclos_60s_interval_combined"]
```
